#Mentoring Question Response - Answer

To compute dot product of numpy nd arrays, you can use numpy.dot() function. This is beacause the numpy.dot() functions accepts two numpy arrays as arguments, computes their dot product and returns the result.

This you have clearly done correctly but you have to understand that the dot product you are trying to calculate is a One dimentional array (1-D array). Hence it should have just one square bracket inside the round brackets for the code to be executed properly.

***Please see example of 1-D array below:***

a = np.array([-3,5,2])

b = np.array([1,2,3])


However, should it have been a Two dimentional arrays (2-D arrays) you still would need to separate each argumaents with a square brackets for each operation and ensure that there are equal values for each set. This will the treated properly once we get there in the coming classes.

***Please see example of 2-D array below:***

a = np.array([[2,3],[3,2]])

b = np.array([[3,2],[5,3]])

```python
#Please see example of 1-D array below:
import numpy as np

a = np.array([-3,5,2])
b = np.array([1,2,3])

np.dot(a,b)
```
